# DifferenceLag

```python
import cudf
import nvtabular as nvt
import numpy as np
import datetime
```

```python
purchases = cudf.DataFrame(
    data={
        'customer_id': np.random.randint(0, 100, 1000),
        'purchase_date': [datetime.date(2022, np.random.randint(1,13), np.random.randint(1,29)) for i in range(1000)],
        'quantity': np.random.randint(1, 50, 1000)
})
purchases.head()
```

```python
purchases = purchases.sort_values(['customer_id', 'purchase_date']).reset_index(drop=True)
purchases.head()
```

```python
ds = nvt.Dataset(purchases)

output = ['quantity'] >> nvt.ops.DifferenceLag(partition_cols=['customer_id'])
wf = nvt.Workflow(output + ['customer_id'])

wf.fit_transform(ds).compute().tail(20)
```
