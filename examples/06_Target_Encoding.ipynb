```python
import nvtabular as nvt
import cudf
import numpy as np

df = cudf.DataFrame(data={
    'cats': list('abbcc'),
    'target': np.random.randint(0,10,5),
})
df
```

```python
out = ['cats'] >> nvt.ops.TargetEncoding('target', kfold=1)

ds = nvt.Dataset(df)
wf = nvt.Workflow(out)

o = wf.fit_transform(ds).compute()
o
```

```python
test = cudf.DataFrame(data={
    'cats': list('abbcc')
})
test
```

```python
test['target'] = 0 # we need to add a dummy variable for now, will likely be removed soon
wf.transform(nvt.Dataset(test)).compute()
```
